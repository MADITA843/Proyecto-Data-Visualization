### 4. Limpieza de datos

Cambio de tipo de dato de object a int64

```python
cols = ["Income","MntWines", "MntFruits", "MntMeatProducts",
    "MntFishProducts", "MntSweetProducts",
    "MntGoldProds", "MntTotal"]
```

Se creó una lista rápida de las columnas que tienen tipo texto, de esta manera hacer un cambio en todas

```python
df1[cols].isna().sum()
```

```python
df1[cols] = (
    df1[cols]
    .replace({r"\$": "", r",": ""}, regex=True)
    .apply(pd.to_numeric, errors="coerce")
)
```

### 6. Reconstrucción de variables

#### Marital

```python
columnas_marital = [
    "marital_Single",
    "marital_Married",
    "marital_Together",
    "marital_Divorced",
    "marital_Widow"]
```

```python
df1["Marital_Status"] = df[columnas_marital].idxmax(axis=1)
```

Como nuestra variable es binaria, usamos el idxmax para que busque el valor máximo, en este caso es el 1.

```python
df1["Marital_Status"] = df["Marital_Status"].str.replace('marital_', '')
```

```python
df1["Marital_Status"].head()
```

```python
columnas_verificacion = [
    "marital_Single",
    "marital_Married",
    "marital_Together",
    "marital_Divorced",
    "marital_Widow",
    "Marital_Status"]
```

```python
df1[columnas_verificacion].sample(10)
```

Como anteriormente se habia mencionado, va a buscar el valor máximo el cual es 1, lo que vendría a ser el estado civil de la persona. Se hizo una verificación de manera aleatoria con 10 personas. Dado esto se procede a eliminar las columnas de estado civil que ya no se trabajaran.

```python
columnas_a_eliminar = [
    "marital_Single",
    "marital_Married",
    "marital_Together",
    "marital_Divorced",
    "marital_Widow"
]
```

```python
df1.drop(columns=columnas_a_eliminar, inplace=True)
```

```python
df1.head()
```

#### Education

```python
columnas_education = [
    "education_Basic",
    "education_Graduation",
    "education_Master",
    "education_PhD"]
```

```python
df1["Education"] = df[columnas_education].idxmax(axis=1)
```

Una vez extraido el nivel de education mediante el indice maximo aprovechando de las columnas binarias procederemos a limpiar a darle formato a la columna education

```python
df1["Education"] = df["Education"].str.replace('education_', '')
```

```python
df1.loc[df[columnas_education].sum(axis=1) == 0, "Education"] = "desconocido"
```

```python
df1.head()
```

Verificamos que las 198 líneas que no tienen un nivel de educación esten identificadas como "desconocido".

```python
df[df1["Education"]=="desconocido"]
```

```python
verificacion_education = [
    "education_Basic",
    "education_Graduation",
    "education_Master",
    "education_PhD",
    "Education"]
```

Verificamos que el nivel de education se haya identificado de forma correcta para luego proceder a eliminar las columnas binarias con las cuales ya no trabajaremos en analisis posteriores.

```python
df1[verificacion_education].sample(10)
```

```python
columnas_a_eliminar_education = [
    "education_Basic",
    "education_Graduation",
    "education_Master",
    "education_PhD"]
```

```python
df1.drop(columns=columnas_a_eliminar_education, inplace=True)
```
